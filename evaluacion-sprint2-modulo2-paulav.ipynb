Si exploramos la columna de state_province veremos que hay universidades cuyo valor para esta columna es None. 
- Cread una función para reemplazar los None por nulos de numpy.

```python
def reemplazar_none (df):
        
    for col in df:
        df[col] = df[col].fillna(np.nan)

    return df
```

- reemplacemos esos nulos por "Unknow". 

```python
def unknow (df):
        
    for col in df:
        df[col] = df[col].fillna("Unknow")

    return df

# Utilizo el método fillna() para reemplazar los valores nulos de cada columna con la cadena "Unknow".
```

Algunos de los valores que tenemos están con siglas, y deberéis reemplazarlos por lo siguiente:
- NV: reemplazalo por Nevada
- TX: reemplazalo por Texas
- IN: reemplazalo por Indianapolis
- CA: reemplazalo por California
- VA: reemplazalo por Virginia
- NY: reemplazalo por New York
- MI: reemplazalo por Michigan
- GA: reemplazalo por Georgia
- ND: reemplazalo por North Dakota

Otros valores que tenemos más formateados son y que deberemos reemplazar:
- New York, NY. Deberéis reemplazarlo por "New York".
- 'Buenos Aires', 'Ciudad Autónoma de Buenos Aires'. En este caso deberéis poner en ambos casos "Buenos Aires" 

```python
def reemplazar_valores (df):
    for col in df:
     df[col] = df[col].replace({"NV":"Nevada",
                                "TX":"Texas",
                                "IN":"Indianapolis",
                                "CA":"California",
                                "VA":"Virginia",
                                "NY":"New York",
                                "MI":"Michigan",
                                "GA":"Georgia",
                                "ND":"North Dakota",
                                "New York, NY":"New York",
                                "Ciudad Autónoma de Buenos Aires":"Buenos Aires"})
     return df
    
# Al ponerlo en formato diccionario uso los ":" y separo cada valor nuevo con una "," en lugar de hacerlo de manera individual con "(" "," ")".
```

```python
reemplazar_valores(df_usa)
reemplazar_valores(df_canada)
reemplazar_valores(df_argentina)
```

Una vez realizados los pasos anteriores, crea una lista con los valores únicos de las provincias de las universidades.

```python
lista_usa = df_usa["state_province"].unique()
lista_canada = df_canada["state_province"].unique()
lista_argentina = df_argentina["state_province"].unique()
```

```python
lista_valores_unicos =list(lista_usa) + list(lista_canada) + list(lista_argentina)
lista_valores_unicos.remove('Unknow')
lista_valores_unicos
```

```python
df_paises = pd.concat([df_usa, df_canada, df_argentina], axis = 0, join= "outer")
df_paises["state_province"].unique()


# Uno los dataframes para usarlos en el siguiente ejercicio.
#join = ‘outer’ se queda todo los datos de todos los dataframes.
#join = ‘inner’ solo se quedan elementos que aparecen en todos los dataframes.
```

Usando la API de geopy, extraed la latitud y la longitud de cada una de las provincias y almacenad los resultados en un dataframe.

```python
df_coords = pd.DataFrame(columns=['state_province', 'latitude', 'longitude'])

for i in lista_valores_unicos:
    location = geolocator.geocode(i)
    df = pd.DataFrame({'state_province': i, 'latitude': location.latitude, 'longitude': location.longitude}, index=[0])
    df_coords = pd.concat([df_coords, df], axis=0, ignore_index=True)

df_coords.head()

# Puse ignore_index=True para que asi los dataframes se coloquen uno encima del otro (verticalmente).
```

Una vez que tengáis los datos del ejercicio anterior en un dataframe, unidlo con el de las universidades que hemos sacado de la API.

```python
df_coords = df_coords.merge(df_paises, how="left", on="state_province").reset_index(drop= True)
df_coords.head()

# Realizo la unión por la columna state_province y como indica el how por la izquierda.
# Añado un reset_index(drop= True) para reiniciar el indice, ya que es posible que la unión haya cambiado el orden y cantidad de filas
```

```python
def crear_insertar_tabla(nombre_bbdd, contraseña, query):
    
    
    cnx = mysql.connector.connect(user='root', password=f"{contraseña}",
                                     host='127.0.0.1', database=f"{nombre_bbdd}",  
                                     auth_plugin = 'mysql_native_password')
    
    mycursor = cnx.cursor()
    
    
    try: 
        mycursor.execute(query)
        cnx.commit() 
    
    except mysql.connector.Error as err:
        print(err)
        print("Error Code:", err.errno)
        print("SQLSTATE", err.sqlstate)
        print("Message", err.msg)

# Esta es la funciones que nos permitirá insertar las tablas        
```

```python
tabla_paises = '''
CREATE TABLE IF NOT EXISTS `universidades`.`paises` (
  `idestado` INT NOT NULL AUTO_INCREMENT,
  `nombre_pais` VARCHAR(45) NULL,
  `nombre_provincia` VARCHAR(45) NULL,
  `latitud` DECIMAL NULL,
  `longitud` DECIMAL NULL,
  PRIMARY KEY (`idestado`))
ENGINE = InnoDB;
'''
```

```python
tabla_universidades = '''
CREATE TABLE IF NOT EXISTS `universidades`.`universidades` (
  `iduniversidades` INT NOT NULL AUTO_INCREMENT,
  `nombre_universidad` VARCHAR(45) NULL,
  `pagina_web` VARCHAR(45) NULL,
  `paises_idestado` INT NOT NULL,
  PRIMARY KEY (`iduniversidades`),
  INDEX `fk_universidades_paises_idx` (`paises_idestado` ASC) VISIBLE,
  CONSTRAINT `fk_universidades_paises`
    FOREIGN KEY (`paises_idestado`)
    REFERENCES `universidades`.`paises` (`idestado`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION)
ENGINE = InnoDB;
'''
# Se especifican las columnas y los tipos de datos.
```

```python
crear_insertar_tabla("universidades","AlumnaAdalab", tabla_paises)
```

```python
crear_insertar_tabla("universidades","AlumnaAdalab", tabla_universidades)

#Se ejecutan las clases y se insertan las tablas
```
